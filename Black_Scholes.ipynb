# Black-Scholes Partial Differential Equation

The Black-Scholes Partial Differential Equation (PDE) is a mathematical equation that plays a pivotal role in options pricing. It was developed by Myron Scholes and Fischer Black in 1973. Their groundbreaking work on options pricing led to their recognition and the award of the Nobel Prize in Economics in 1997.

## Assumptions

To derive the Black-Scholes PDE and its solution, the model makes several key assumptions about the market conditions and underlying assets. These assumptions are as follows:

1. **Constant Short Term Interest Rates:** The model assumes that short-term interest rates are constant throughout the options' lifespan. This means the risk-free rate remains the same over time.

2. **No Dividends on Stocks:** The model assumes that the underlying stock does not pay any dividends during the options' lifespan. This simplification is made to avoid complications related to dividend payments.

3. **No Transaction Costs:** The Black-Scholes model assumes there are no transaction costs associated with buying or selling the underlying asset or the options themselves. This idealization is used to make the calculations more manageable.

4. **Ability to Borrow and Lend:** Traders are assumed to have the ability to borrow and lend money at the risk-free rate. This implies that there are no restrictions on short selling, and investors can borrow or lend any fraction of the stock.

5. **Short Selling Allowed:** The model allows for short selling, which means traders can sell assets they do not currently own, with the intention of buying them back at a later time.

It's important to note that these assumptions are idealized and do not necessarily represent real-world market conditions. Nevertheless, they provide a strong foundation for understanding and pricing options in a simplified and controlled setting.

The Black-Scholes PDE is the cornerstone of modern option pricing theory, and its solutions have been instrumental in the development of financial markets and derivative instruments.


## Derivation Method Overview for Call Option Pricing using Black-Scholes Model

The Black-Scholes model is a widely-used mathematical framework for pricing European call options. It is based on the concept of dynamic replication and uses the concept of a risk-free portfolio to derive the famous Black-Scholes Partial Differential Equation (PDE). This PDE is then solved to obtain the call option's theoretical price.

Let's walk through the derivation method step-by-step:

1. **Price Derivative Using Replication:**
The first step involves considering the call option's price as a derivative security. We aim to replicate the payoff of the call option using a combination of the underlying stock and a risk-free asset. By replicating the call option's payoffs, we create a portfolio that behaves similarly to holding the call option itself.

2. **Construct Risk-Free Portfolio:**
To replicate the call option, we construct a portfolio using the underlying stock and the risk-free asset (e.g., a money market account). The portfolio's value must be risk-free, meaning it is not subject to uncertainty or market risk. The risk-free portfolio acts as a benchmark against which the call option's price is determined.

3. **Assuming Our Call Option is a Smooth Function Over All $C$:**
In this step, we assume that the call option's price function, denoted as $C$ {$C_t = C(S, t)$}, is a smooth function for all possible values of the underlying stock price ($S$) and time ($t$). The smoothness assumption simplifies the mathematical analysis.

4. **Use Ito's Rule to Express Portfolio Drift as $C$'s Partial Derivatives:**
Ito's lemma is a powerful tool used in stochastic calculus to handle stochastic processes. In the context of the Black-Scholes model, we use Ito's rule to express the change in the portfolio's value (the risk-free portfolio) as the partial derivatives of the call option's price function, $C$, with respect to the underlying stock price ($S$) and time ($t$). This step links the dynamics of the portfolio with the call option's price function.

5. **Find $C(S, t)$ That Satisfies the Partial Differential Equation:**
By equating the drift expression obtained in the previous step with the risk-free rate of return, we derive the Black-Scholes Partial Differential Equation (PDE). The PDE is a second-order partial differential equation involving the call option price function, $C$, the underlying stock price ($S$), and time ($t$). We then solve this PDE subject to appropriate boundary conditions to find the closed-form solution for the call option's price, $C(S, t)$.

The final solution for $C(S, t)$ represents the theoretical price of a European call option under the assumptions of the Black-Scholes model. It provides a powerful tool for pricing and understanding options in ideal market conditions with constant short-term interest rates, no dividends, no transaction costs, and the ability to borrow and lend at the risk-free rate.

Please note that while the Black-Scholes model has been historically influential, real-world market conditions may not always align with its assumptions, and market participants should exercise caution and adapt the model to suit specific situations.


### Construct Portfolio with Option and Stocks

In this section, we will construct a portfolio consisting of one call option and $-\alpha_t$ shares of the underlying stock. We will denote this portfolio as $V_t$ and then derive its dynamics.

1. The portfolio $V_t$ is constructed as follows:

$$    V_t = C_t - \alpha S_t$$
$$dV_t = dC_t - \alpha dS_t$$

Here, $C_t$ represents the price of the call option at time $t$, and $S_t$ represents the price of the underlying stock at time $t$. The parameter $\alpha$ denotes the number of shares held in the portfolio, and the negative sign indicates that the portfolio has a short position in the stock.

2. Assuming the underlying stock follows the Geometric Brownian Motion (GBM):

$$dS_t = \mu S_t dt + \sigma S_t dW_t$$

Where:
- $dS_t$ is the change in the stock price over a small time interval $dt$.
- $S_t$ is the current stock price at time $t$.
- $\mu$ is the drift term (expected return of the stock per unit time).
- $\sigma$ is the constant volatility (standard deviation of the stock return per unit time).
- $dW_t$ is a Wiener process (Brownian motion) representing random movements in the stock price.

3. Deriving the call price:
The call price, $C_t$, is dependent on the underlying stock price ($S_t$) and time ($t$), i.e., $C_t = C(S_t, t)$. We can find the partial derivatives of $C_t$ with respect to each parameter as follows:

    $$\partial C_t = \frac{\partial C_t}{\partial t}dt + \frac{\partial C_t}{\partial S}dS + \frac{1}{2}\left(\frac{\partial^2C}{\partial S^2}\partial S^2 + \frac{\partial^2C}{\partial t^2}\partial t^2\right)$$
    
Since the time squared terms are negligible, we substitute the derivative of this value back into the derivative of $C_t$:

$$\partial V_t = \frac{\partial C_t}{\partial t}dt + \frac{\partial C_t}{\partial S}dS + \frac{1}{2} \frac{\partial^2C}{\partial S^2} \partial S^2 - \alpha dS_t $$

    
    
4. Assuming $\alpha = \frac{\partial C_t}{\partial S}$ (number of shares held in the portfolio), this will cancel some terms in the equation.

5. Using GBM, the term $dS_t^2$ is calculated as:

    $$dS_t^2 = (\mu S_t dt + \sigma S_t dW)^2$$
    $$dS_t^2 = \sigma^2 S_t^2 dt$$


6. Creating an Arbitrage-Free Portfolio with the Bank Account:
Now, let's create an arbitrage-free portfolio by including the bank account with a constant rate $r$.

    $$dV_t = rV_tdt$$

Where $dB_t = rB_tdt$, and $B_t$ represents the bank account balance.

7. Equating both sides and rearranging, we get the Black-Scholes PDE for a European call option:

    $$\boxed{\frac{\partial C_t}{\partial t} dt + \frac{1}{2} \frac{\partial^2C}{\partial S^2} \alpha^2 S_t^2 + r S_t \frac{\partial C_t}{\partial S} = rC_t}$$
    

Boundary Conditions:
At expiry, the payoff of the call option is given by:

$$\boxed{C(S,t)=(S-K)^{+}=max\{0, S-K\}}$$
    
Where $K$ is the strike price of the call option.

The Black-Scholes PDE, along with these boundary conditions, is used to solve for the theoretical price of the European call option, taking into account the ideal market assumptions of constant short-term interest rates, no dividends, no transaction costs, and continuous trading without restrictions on short selling and borrowing.


# Black-Scholes Option Pricing Formula in Python
The Black-Scholes model, developed by Fischer Black, Myron Scholes, and Robert Merton in the early 1970s, is a mathematical formula widely used in finance to estimate the theoretical price of European-style call and put options. It assumes ideal market conditions, including continuous trading, no transaction costs, constant short-term interest rates, and the absence of dividends.

The formula calculates the option price by considering several factors, including the risk-free interest rate, underlying asset price, strike price, time to expiration, and volatility of the underlying asset's returns.

### Import Libraries
To implement the Black-Scholes formula, we need to import the necessary libraries. We use 'numpy' for mathematical calculations and 'scipy.stats' to access the standard normal cumulative distribution function (CDF) required for the option price computation.

### Define Function: blackScholes
Next, we define the 'blackScholes' function, which calculates the Black-Scholes price of call or put options. 

### Calculate Option Price
We begin by computing the two parts of the Black-Scholes formula: 'd1' and 'd2'. These terms are used to estimate the probabilities of the option expiring in-the-money or out-of-the-money.

### Compute Option Price
Next, we calculate the option price using the computed 'd1' and 'd2'. The specific formula depends on the option type (call or put).

For a call option, the price is calculated as 'S * norm.cdf(d1, 0, 1) - K * np.exp(-r * T) * norm.cdf(d2, 0, 1)'.
For a put option, the price is calculated as 'K * np.exp(-r * T) * norm.cdf(-d2, 0, 1) - S * norm.cdf(-d1, 0, 1)'.

### Handle Invalid Option Type
We include error handling in case an invalid option type is provided.

```python
# Implementation the black-scholes formula in python

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use('seaborn')

def blackScholes(r, S, K, T, sigma, option_type="c"):
    """
    Calculate the Black-Scholes price of a call or put option.

    Parameters:
    r : float
        The risk-free interest rate.
    S : float
        The current price of the underlying asset (stock price).
    K : float
        The strike price (exercise price) of the option.
    T : float
        The time to expiration in years.
    sigma : float
        The annualized volatility of the underlying asset's returns.
    option_type : str, optional
        Type of option to be priced, 'c' for call (default) and 'p' for put.

    Returns:
    float
        The option price according to the Black-Scholes formula.
    """
    
    d1 = (np.log(S/K) + (r + (sigma**2)/2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == "c":
        price = S * norm.cdf(d1, 0, 1) - K * np.exp(-r * T) * norm.cdf(d2, 0, 1)
    elif option_type == "p":
        price = K * np.exp(-r * T) * norm.cdf(-d2, 0, 1) - S * norm.cdf(-d1, 0, 1)
    else:
        raise ValueError("Invalid option type. Please use 'c' for Call or 'p' for Put.")
    
    return price
```

### Example Usage
Finally, we provide an example of how to use the 'blackScholes' function to calculate the option price for a call option.

```python
# Example usage
r = 0.01
S = 30
K = 40
T = 240/365
sigma = 0.30
option_price = blackScholes(r, S, K, T, sigma, option_type="c")
print("Call Option Price: ", option_price)
```

### Additional Examples
You can use the 'blackScholes' function to calculate the option price for various scenarios by changing the input values of 'S', 'K', 'T', and 'sigma'. Below are additional examples:

```python
# Additional examples
# Put option pricing
put_option_price = blackScholes(r, S, K, T, sigma, option_type="p")
print("Put Option Price: ", round(put_option_price, 3))

# Different stock prices
S_2 = 45
call_option_price_2 = blackScholes(r, S_2, K, T, sigma, option_type="c")
print("Call Option Price with S=45: ", round(call_option_price_2, 3))

# Different strike prices
K_2 = 35
call_option_price_3 = blackScholes(r, S, K_2, T, sigma, option_type="c")
print("Call Option Price with K=35: ", round(call_option_price_3, 3))

# Different volatilities
sigma_2 = 0.25
call_option_price_4 = blackScholes(r, S, K, T, sigma_2, option_type="c")
print("Call Option Price with sigma=0.25: ", round(call_option_price_4, 3))
```

Using different combinations of S, K, T, and sigma, you can explore how the option price varies under different market conditions. Additionally, you can calculate put option prices or even compare the prices for different strike prices and volatilities.

```python
# Plot Option Prices against Underlying Asset Prices
underlying_asset_prices = np.linspace(20, 50, 100)
call_option_prices = [blackScholes(r, S, K, T, sigma, option_type="c") for S in underlying_asset_prices]
plt.plot(underlying_asset_prices, call_option_prices)
plt.xlabel("Underlying Asset Price (S)")
plt.ylabel("Call Option Price")
plt.title("Call Option Price vs. Underlying Asset Price")
plt.grid(True)
plt.show()

# Predict Option Price for Different Strike Prices
strike_prices = np.linspace(20, 50, 100)
call_option_prices_strikes = [blackScholes(r, S, K, T, sigma, option_type="c") for K in strike_prices]
plt.plot(strike_prices, call_option_prices_strikes)
plt.xlabel("Strike Price (K)")
plt.ylabel("Call Option Price")
plt.title("Call Option Price vs. Strike Price")
plt.grid(True)
plt.show()

# Predict Option Price for Different Time to Expirations
time_to_expirations = np.linspace(1, 180, 100) / 365
call_option_prices_time = [blackScholes(r, S, K, T, sigma, option_type="c") for T in time_to_expirations]
plt.plot(time_to_expirations * 365, call_option_prices_time)
plt.xlabel("Time to Expiration (Days)")
plt.ylabel("Call Option Price")
plt.title("Call Option Price vs. Time to Expiration")
plt.grid(True)
plt.show()
```
